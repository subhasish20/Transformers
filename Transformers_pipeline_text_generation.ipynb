```python
# transformers pipeline
from transformers import pipeline
```

```python
generator = pipeline("text-generation", model="gpt2")
```

```python

output = generator("Deep learning is", max_length=10, num_return_sequences=1)
```

```python
print(output[0]['generated_text'])
```
